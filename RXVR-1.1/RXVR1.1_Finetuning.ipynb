```python
import subprocess

# Define input/output paths and pretrained checkpoint
input_path = "SE000002.nii"                     # Input CT volume
output_path = "output_finetuned"                # Output directory for finetuned model
ckpt_path = "output\RXVR1.1_best.pth"           # Path to pretrained checkpoint

command = [
    "xvr", "finetune",
    "--inpath", input_path,
    "--outpath", output_path,
    "--ckptpath", ckpt_path,
    "--lr", "0.005",                            # Optional: learning rate
    "--batch_size", "116",
    "--n_epochs", "10",
    "--n_batches_per_epoch", "25",
    "--rescale", "1.0",
    "--project", "RXVR1.1",                     # Optional: wandb project name
    "--name", "RXVR1.1_Finetuned"
]

# Run the command and stream output
process = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)

for line in process.stdout:
    print(line, end='')
```
